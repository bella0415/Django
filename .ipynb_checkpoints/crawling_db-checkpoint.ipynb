## 등급 표시

- 용감한맛 쿠키 ~ 보더맛 쿠키 common
- 파일럿맛 쿠키 ~ 치어리더맛 쿠키 rare
- 에일리언 도넛맛 쿠키 ~ 아몬드맛 쿠키 epic
- 어둠 마녀 쿠키 ~ 불꽃정령 쿠키 lengendary
- 헬로키티 ~ 쿠키 로이드 special

```python
ovenbreak_cookie =[]
for i in range(len(oven_name_list)) :
    name = oven_name_list[i]
    gender = oven_gender_list[i]
    skill = oven_skill_list[i]
    pet = oven_pet_list[i]
    candy = oven_candy_list[i]
    img = oven_img_list[i]
    pet_img = oven_pet_img_list[i]
   
    if i < oven_name_list.index("파일럿맛 쿠키"):
        rate = 'common'
    elif i < oven_name_list.index("에일리언 도넛") :
        rate = 'rare'
    elif i < oven_name_list.index("어둠마녀 쿠키") :
        rate = 'epic'
    elif i < oven_name_list.index("헬로키티") :
        rate = 'lengendary'
    elif i < len(oven_name_list) :
        rate = 'special'
    
    cookie_info = f"{name}\t{gender}\t{skill}\t{pet}\t{candy}\t{rate}\t{img}\t{pet_img}"
    dic = {'name' : name,'gender' : gender,'skill' : skill,'pet' : pet,'candy' : candy, 'rate' : rate,'img' : img, 'pet_img' : pet_img}
    ovenbreak_cookie.append(dic)
print(len(ovenbreak_cookie),ovenbreak_cookie)
```

# pymysql vs sqlalchemy
- pymysql 
    - 정교하게 할 수 있음
    - 별도의 commit 필요

- sqlalchemy 
    - 프로그램에서 다루기 쉬운 구조체 방식씀
    - 별도의 commit 필요 x (사용 후 적용 해주면 됨)
    - 빠름, 단일 쿼리 사용 좋음
    - db 종류와 상관없이 코드 일관
